預測並取最多的為結果

```python
import cv2
import tensorflow as tf
import numpy as np
import os
import skimage.io

CATEGORIES = ["Coarse speckled", "Cytoplasmic", "Diffused", "Discrete speckled", "Fine speckled",
          "Nucleolar", "Peripheral"]
count=[0,0,0,0,0,0,0]

def prepare(file):
    IMG_SIZE = 50
    img_array = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)

model = tf.keras.models.load_model("CNN.model_3")

IMAGE_DIR="E:\\together\\tiffany\\Mask_RCNN-master\\samples\\nucleus_crop\\datasets\\ANA_test_crop_datas"
file_names = next(os.walk(IMAGE_DIR))[2]
#資料夾內圖片數量
quantity=(len([name for name in os.listdir(IMAGE_DIR) if os.path.isfile(os.path.join(IMAGE_DIR, name))]))

for i in range(0, quantity):
    image = IMAGE_DIR+ '\\' + file_names[i]
    #print(image)
    
    #image=str(image)
    #print(type(image))
    image=prepare(image)



    prediction = model.predict([image])
    #print(prediction)
    prediction = list(prediction[0])
    print(CATEGORIES[prediction.index(max(prediction))])
    
    #預測出的那種細胞, 那種細胞加一
    count[prediction.index(max(prediction))]=count[prediction.index(max(prediction))]+1
    
    i=i+1

#印出每種預測出的數量
print(count)
#印出最多的那種
print(max(count))
#預測結果 aka 印出最多的那種對照的種類名稱
print(CATEGORIES[count.index(max(count))])

#正確答案
```
